# 🌐 International Labour Organization (ILO)

La **Organización Internacional del Trabajo (OIT)** es una agencia especializada de las Naciones Unidas
fundada en 1919. Es la única agencia de la ONU con estructura tripartita: gobiernos, empleadores y
trabajadores tienen representación y voz en sus decisiones.

## Misión
Promover el trabajo decente y productivo en condiciones de libertad, equidad, seguridad y dignidad humana.

## Relevancia para este análisis
La OIT es la fuente de referencia global para estadísticas laborales. Sus datos alimentan directamente
los indicadores de los Objetivos de Desarrollo Sostenible (ODS) relacionados con trabajo, entre ellos:

| ODS | Indicador |
|-----|-----------|
| 5.5.2 | Proporción de mujeres en puestos directivos |
| 8.3.1 | Proporción de empleo informal por sexo |
| 8.5.1 | Ingresos medios por hora por sexo |
| 8.5.2 | Tasa de desempleo por sexo y edad |
| 8.6.1 | Proporción de jóvenes NEET por sexo |

## Fuente de datos
Los datos utilizados en este notebook provienen de **ILOSTAT**, la base de datos estadísticos de la OIT,
accesible via API en `https://rshiny.ilo.org/dataexplorer01/`.

> 📎 Sitio oficial: [ilo.org](https://www.ilo.org) | Datos: [ilostat.ilo.org](https://ilostat.ilo.org)

```python
# @title
import pandas as pd
import requests
from io import StringIO

def read_csv(url):
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers)
    response.raise_for_status()

    df = pd.read_csv(StringIO(response.text))
    return df
```

```python
from tabulate import tabulate

def summarize_by_country(df, countries, country_col='ref_area.label', time_col='time'):
    rows = []
    for country in countries:
        aux = df[df[country_col] == country]
        rows.append([country, f"{len(aux):,}", aux[time_col].min(), aux[time_col].max()])
    print(tabulate(rows, headers=["País", "Registros", "Desde", "Hasta"], tablefmt="pretty"))

summarize_by_country(df, latam_countries)
```

# 🌎 Comisión Económica para América Latina y el Caribe (CEPAL)

La **CEPAL** es una de las cinco comisiones regionales de las Naciones Unidas, fundada en 1948
con sede en Santiago de Chile. Su misión es contribuir al desarrollo económico y social de
América Latina y el Caribe mediante investigación, análisis y asistencia técnica a los países
de la región.

## Misión
Promover el desarrollo económico y social de América Latina y el Caribe a través de la
cooperación y la integración regional.

## CEPALSTAT
**CEPALSTAT** es el portal de estadísticas de la CEPAL, que centraliza indicadores económicos,
sociales y ambientales de los países de la región. Sus datos provienen de fuentes nacionales,
organismos internacionales (como la OIT, OMS, UNESCO) y producción propia a través de
**CELADE** (Centro Latinoamericano y Caribeño de Demografía).

## Relevancia para este análisis
CEPALSTAT ofrece mayor granularidad regional que otras fuentes globales, con indicadores
específicos para América Latina y el Caribe en las siguientes áreas:

| Área | Ejemplos |
|------|---------|
| Demografía | Población por sexo, edad y país |
| Género | Participación laboral, violencia, autonomía |
| Trabajo | Empleo, informalidad, brechas salariales |
| ODS | Seguimiento de los 17 objetivos para la región |

## Diferencia con ILO/ILOSTAT

| | ILO | CEPALSTAT |
|---|---|---|
| Cobertura | Global | América Latina y el Caribe |
| Foco | Estándares y mercado laboral | Social, económico y ambiental |
| Demografía | Limitada | CELADE (proyecciones 1950-2100) |
| Granularidad regional | Media | Alta |

## Fuente de datos
Los datos utilizados en este notebook provienen de la **API pública de CEPALSTAT**,
accesible en `https://statistics.cepal.org/portal/cepalstat/open-data.html?lang=es#/`.

> 📎 Sitio oficial: [cepal.org](https://www.cepal.org) | Datos: [estadisticas.cepal.org](https://estadisticas.cepal.org)

```python
# @title
import requests
from rich import print_json
import json

def request_api(api):
    response = requests.get(api)
    response.raise_for_status()
    return response.json()

api = "https://api-cepalstat.cepal.org/cepalstat/api/v1/thematic-tree?lang=en&format=json"
data = request_api(api)
```

### Menu Principal de la API

```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output, JSON

selected_indicator = {'id': None, 'name': None}
out = widgets.Output()
box = widgets.HBox([])

def make_options(children):
    return [(f"[{i}] {x['name']}", i) for i, x in enumerate(children)]

def rebuild(level, indices):
    current_selects = list(box.children)[:level]

    try:
        node = data['body']
        for idx in indices:
            node = node['children'][idx]

        children = node.get('children', [])

        if not children:
            with out:
                clear_output()
                selected_indicator['id']   = node.get('indicator_id')
                selected_indicator['name'] = node.get('name')
                display(JSON(node))
        else:
            with out:
                clear_output()

        if children:
            new_sel = widgets.Select(
                options=make_options(children),
                description=f'Nivel {level}:',
                rows=8,
                layout=widgets.Layout(width='300px')
            )

            def make_observer(lvl, idxs):
                def observer(change):
                    if change['new'] is not None:
                        rebuild(lvl + 1, idxs + [change['new']])
                return observer

            new_sel.observe(make_observer(level, indices[:]), names='value')
            current_selects.append(new_sel)
            box.children = tuple(current_selects)

            rebuild(level + 1, indices + [0])
            return

        box.children = tuple(current_selects)

    except (IndexError, KeyError):
        box.children = tuple(current_selects)

# nivel 0
root_children = data['body'].get('children', [])
sel0 = widgets.Select(
    options=make_options(root_children),
    description='Nivel 0:',
    rows=8,
    layout=widgets.Layout(width='300px')
)

def on_sel0(change):
    if change['new'] is not None:
        rebuild(1, [change['new']])

sel0.observe(on_sel0, names='value')
box.children = (sel0,)

display(box, out)

# expande desde el primer item
rebuild(1, [0])
```

```python
# @title
def build_dim_lookup(dimensions):
    """
    Retorna un dict: { 'dim_{id}': { member_id: member_name } }
    """
    lookup = {}
    for dim in dimensions:
        key = f"dim_{dim['id']}"
        lookup[key] = {m['id']: m['name'] for m in dim['members']}
    return lookup

indicator_id = selected_indicator['id']

# Indicadores
api_ind = f"https://api-cepalstat.cepal.org/cepalstat/api/v1/indicator/{indicator_id}/records?lang=en&format=json"
data_ind = request_api(api_ind)
df_ind = pd.DataFrame(data_ind['body']['data'])

#Dimensiones
api_dim = f"https://api-cepalstat.cepal.org/cepalstat/api/v1/indicator/{indicator_id}/dimensions?lang=en&format=json"
dim_data = request_api(api_dim)
dimensions = dim_data['body']['dimensions']
lookup = build_dim_lookup(dimensions)

#Mappeamos codigos
for col, mapping in lookup.items():
    if col in df_ind.columns:
        df_ind[col] = df_ind[col].map(mapping)

#Renombramos columnas
col_names = {f"dim_{dim['id']}": dim['name'].replace("__ESTANDAR", "").lower() for dim in dimensions}
df_ind.rename(columns=col_names, inplace=True)

# metadata del indicador
api_sources = f"https://api-cepalstat.cepal.org/cepalstat/api/v1/indicator/{indicator_id}/sources?lang=en&format=json"
api_notes   = f"https://api-cepalstat.cepal.org/cepalstat/api/v1/indicator/{indicator_id}/footnotes?lang=en&format=json"

sources = request_api(api_sources)
notes   = request_api(api_notes)

source_lookup = {s['id']: s['description'] for s in sources['body']['sources']}
notes_lookup = {str(n['id']): n['description'] for n in notes['body']['footnotes']}

df_ind['source_id']  = df_ind['source_id'].map(source_lookup)
df_ind['notes_ids']  = df_ind['notes_ids'].map(notes_lookup)

df_ind.rename(columns={'source_id': 'source', 'notes_ids': 'note'}, inplace=True)
df_ind.head()
```

#### Metadata

```python
# @title
from IPython.display import Markdown
from bs4 import BeautifulSoup
import re

def clean_methodology(text):
    soup = BeautifulSoup(text, 'html.parser')

    # detecta CUALQUIER fraccion con patron border-bottom
    for container in soup.find_all('div', style=lambda s: s and 'text-align: center' in s):
        fractions = []

        for inline in container.find_all('div', style=lambda s: s and 'inline-block' in s):
            # prefijo: span antes de la fraccion (ej: "yPC = ")
            prefix = ''
            prefix_span = inline.find('span')
            if prefix_span:
                prefix = prefix_span.get_text().strip()

            # numerador: div con border-bottom
            num_div = inline.find('div', style=lambda s: s and 'border-bottom' in s)
            if not num_div:
                continue

            # denominador: siguiente sibling
            den_div = num_div.find_next_sibling('div')
            if not den_div:
                continue

            num = num_div.get_text().strip()
            den = den_div.get_text().strip()

            latex = f'{prefix}\\frac{{{num}}}{{{den}}}'
            fractions.append(latex)

        if fractions:
            container.replace_with(f'\n\n$${"  ".join(fractions)}$$\n\n')

    # listas
    for li in soup.find_all('li'):
        li.replace_with(f'\n- {li.get_text().strip()}')

    # parrafos
    for p in soup.find_all('p'):
        p.replace_with(f'\n\n{p.get_text().strip()}')

    result = soup.get_text()
    result = re.sub(r'\n{3,}', '\n\n', result)
    result = re.sub(r'[ \t]{2,}', ' ', result)
    return result.strip()

def build_indicator_markdown(m, dimensions, sources, notes):

    # dimensiones
    dim_lines = []
    for dim in dimensions:
        members = [mb['name'] for mb in dim['members'] if mb.get('in') == 1]
        dim_lines.append(f"- **{dim['name']}** ({len(members)} valores activos): {', '.join(members[:5])}{'...' if len(members) > 5 else ''}")

    # fuentes
    source_lines = [f"- {s['description']} ([{s['organization_acronym']}]({s['organization_url']}))"
                    for s in sources['body']['sources']]

    # notas
    note_lines = [f"- `{n['id']}`: {n['description'][:120]}..." for n in notes['body']['footnotes']]

    md = f"""# {m['indicator_name']}

## Overview
| Campo | Valor |
|---|---|
| **ID** | {m['indicator_id']} |
| **Tema** | {m['theme']} |
| **Área** | {m['area']} |
| **Unidad** | {m['unit']} |
| **Última actualización** | {m['last_update']} |
| **Decimales** | {m['decimals']} |

## Definición
{m['definition']}

## Cobertura
{m['data_features']}

## Dimensiones
{chr(10).join(dim_lines)}

## Metodología
{clean_methodology(m['calculation_methodology'])}

## Fuentes
{chr(10).join(source_lines)}

## Notas
{chr(10).join(note_lines)}

## Comentarios
{m['comments']}
"""
    return md

api_meta = f"https://api-cepalstat.cepal.org/cepalstat/api/v1/indicator/{indicator_id}/metadata?lang=en&format=json"

meta = request_api(api_meta)
meta = meta['body']['metadata']

md = build_indicator_markdown(meta, dimensions, sources, notes)

display(Markdown(md))
```

# 🌐 Unión Internacional de Telecomunicaciones (ITU) via Banco Mundial

La **ITU** (International Telecommunication Union) es el organismo especializado de las
Naciones Unidas para las tecnologías de la información y la comunicación (TIC), fundado
en 1865 con sede en Ginebra, Suiza. Es la principal fuente global de estadísticas sobre
conectividad y uso de internet desagregadas por sexo, edad y país.

## Misión
Facilitar la conectividad internacional en redes de comunicación y garantizar que las TIC
sean accesibles para todos, promoviendo el desarrollo digital inclusivo a nivel global.

## Datos utilizados

Los indicadores de uso de internet desagregados por sexo provienen de la **API pública
del Banco Mundial**, que replica y distribuye los datos oficiales de la ITU:

| Indicador | Código | Descripción |
|-----------|--------|-------------|
| Usuarios de internet (total) | `IT.NET.USER.ZS` | % de población total que usa internet |
| Usuarios de internet (mujeres) | `IT.NET.USER.FE.ZS` | % de mujeres que usan internet |
| Usuarios de internet (hombres) | `IT.NET.USER.MA.ZS` | % de hombres que usan internet |

## Relevancia para este análisis

La brecha digital de género se calcula como la diferencia entre el porcentaje de hombres
y mujeres que usan internet por país y año. Esta variable es uno de los dos pilares del
índice de riesgo del dashboard, junto con la carga de cuidados.

| Área | Detalle |
|------|---------|
| Cobertura geográfica | 23 países de América Latina y el Caribe |
| Cobertura temporal | 2010 - 2023 |
| Granularidad | Nacional, desagregado por sexo |
| ODS relacionado | ODS 5.b — Acceso a TIC para mujeres |

## Diferencia entre fuentes

| | ITU directo | Banco Mundial (`wbgapi`) |
|---|---|---|
| Acceso | Descarga manual (Excel) | API REST programática |
| Desagregación por sexo | Sí | Sí |
| Actualización | Anual | Anual |
| Integración en Python | Scraping | `wbgapi` nativo |


> 📎 ITU: [itu.int](https://www.itu.int) | Banco Mundial: [data.worldbank.org](https://data.worldbank.org/indicator)
