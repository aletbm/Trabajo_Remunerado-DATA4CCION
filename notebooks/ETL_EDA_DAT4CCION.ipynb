## 1\.1\.2\. Transformación

```
df = internet.copy()

# Limpiar columnas
for c in ['SEX_LABEL', 'AGE_LABEL', 'REF_AREA_LABEL', 'INDICATOR_LABEL']:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip()

# Convertir año
df['TIME_PERIOD'] = pd.to_numeric(df['TIME_PERIOD'], errors='coerce').astype('Int64')

# Eliminar nulos clave
df = df.dropna(subset=['SEX_LABEL', 'REF_AREA_LABEL', 'TIME_PERIOD'])

# Países correctos
paises_objetivo = [
    'Argentina', 'Colombia', 'Dominican Republic', 'Costa Rica',
    'Guatemala', 'Uruguay', 'Chile', 'El Salvador', 'Peru'
]

# Filtros
mask_mujeres = df['SEX_LABEL'] == 'Female'
mask_anios = df['TIME_PERIOD'].between(2020, 2025)
mask_paises = df['REF_AREA_LABEL'].isin(paises_objetivo)
# Aplicar filtros principales
filtrado = df[mask_mujeres & mask_anios & mask_paises]

# QUEDARSE SOLO con el total (sin desglose por edad)
filtrado = filtrado[
    filtrado['AGE_LABEL'] == 'All age ranges or no breakdown by age'
]

# Seleccionar columnas incluyendo OBS_VALUE
columnas = ['REF_AREA_LABEL', 'TIME_PERIOD', 'AGE_LABEL', 'OBS_VALUE']
columnas = [col for col in columnas if col in filtrado.columns]

resultado = filtrado[columnas].sort_values(
    ['REF_AREA_LABEL', 'TIME_PERIOD']
)

print("Filas filtradas:", len(resultado))
display(resultado)
```

```
print(sorted(resultado['REF_AREA_LABEL'].unique()))
```

```
# Obtener el último año por país (tabla pequeña)
ultimos_anios = resultado.groupby('REF_AREA_LABEL', as_index=False)['TIME_PERIOD'].max()

# Unir con los datos originales
resultado_final = resultado.merge(
    ultimos_anios,
    on=['REF_AREA_LABEL', 'TIME_PERIOD'],
    how='inner'
)

# Ordenar
resultado_final = resultado_final.sort_values(
    ['REF_AREA_LABEL', 'AGE_LABEL']
)

display(resultado_final)
```

```
# Normalización de nombres de columnas para mejorar la legibilidad
resultado_final = resultado_final.rename(
    columns=
    {'REF_AREA_LABEL': 'pais',
    'TIME_PERIOD': 'anio',
    'AGE_LABEL': 'edad',
    'OBS_VALUE': 'porcentaje_uso_internet'}
)

# Mostrar el resultado
resultado_final.head(10)
```

```
# Definimos diccionario de reemplazos al español
dicc_paises= {
    'Argentina': 'Argentina',
    'Chile': 'Chile',
    'Colombia': 'Colombia',
    'Costa Rica': 'Costa Rica',
    'Dominican Republic': 'Republica Dominicana', 
    'El Salvador': 'El Salvador',
    'Guatemala': 'Guatemala',
    'Peru': 'Peru',
    'Uruguay': 'Uruguay'
    }

# Aplicamos el reemplazo en la columna 'pais'
resultado_final['pais'] = resultado_final['pais'].replace(dicc_paises)

# Verificación de resultados
print("Países procesados:")
print(resultado_final['pais'].unique())

resultado_final.head(10)
```

```
# Estandarización de la columna 'pais' para garantizar consistencia en la clave de unión.
# Se eliminan espacios, se unifica el formato en mayúsculas y se remueven acentos.
resultado_final['pais'] = (
    resultado_final['pais']
    .str.strip()
    .str.upper()
    .str.normalize('NFKD')
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
)

# Mostrar resultados
resultado_final['pais'].unique()
```

# 2\. Análisis Exploratorio de Datos \(EDA\)

En esta etapa realizamos un análisis exploratorio del dataset con el objetivo de comprender la estructura de los datos y preparar la información para la construcción del mapa interactivo\.

El EDA se enfoca en tres ejes principales:

\- Distribución poblacional femenina por país en edad activa, para identificar dónde se concentra la población con mayor potencial de participación en el trabajo remunerado\.
\- Carga de trabajo \(remunerado y no remunerado\), con especial énfasis en la proporción de trabajo no remunerado como indicador de carga de cuidados\.
\- Brecha digital, analizando el nivel de uso de internet como aproximación al acceso a conectividad\.

A partir de estos ejes, se construyen indicadores que permiten resumir la información:

\- Carga de cuidados  
\- Brecha digital  
\- Vulnerabilidad estructural  
\- Peso poblacional en edad activa  
\- Impacto total  

El objetivo del EDA es identificar patrones entre países y generar los insumos necesarios para el desarrollo del índice y el mapa interactivo final\.

## 2\.1\. Carga de cuidados

Mide la proporción del tiempo total de trabajo femenino que se destina a trabajo no remunerado \(cuidados y tareas domésticas\)\.

Cálculo: tiempo de trabajo no remunerado / tiempo total de trabajo

```
import numpy as np

df_tiempo_internet["carga_cuidados"] = (
    df_tiempo_internet["tiempo_trabajo_no_remunerado"] /
    df_tiempo_internet["tiempo_total_trabajo"].replace(0, np.nan)
) * 100
```

```
# Revisamos las primeras filas
df_tiempo_internet[
    ["pais",
    "tiempo_total_trabajo", "tiempo_trabajo_no_remunerado",
    "carga_cuidados"]
].head(10)
```

```
# Ranking de carga de cuidados por país en mujeres
df_tiempo_internet[["pais", "carga_cuidados"]].sort_values(
    by="carga_cuidados",
    ascending=False
)
```

Conclusión preliminar: Observamos que Guatemala es el país, dentro de la muestra analizada, donde las mujeres presentan una mayor carga de cuidados, dedicando un 74\.42% de su tiempo de trabajo a tareas no remuneradas\. Le sigue El Salvador con un 67\.49%\.

## 2\.2\. Brecha digital

Mide el nivel de acceso faltante a internet en la población femenina, como aproximación a la brecha de conectividad digital\. Los valores más altos indican menor nivel de acceso y mayor brecha digital\. Por ejemplo, si el 70% de la población usa internet, la brecha digital es del 30%\.

Cálculo: 100 \- % uso de internet

```
df_tiempo_internet["brecha_digital"] = 100 - df_tiempo_internet["porcentaje_uso_internet"]
```

```
# Revisamos las primeras filas
df_tiempo_internet[["pais", "porcentaje_uso_internet", "brecha_digital"]].head(10)
```

```
# Ranking de uso de internet por país en mujeres
df_tiempo_internet[["pais", "brecha_digital"]].sort_values(
    by="brecha_digital",
    ascending=False
)
```

Conclusión preliminar: Se observa una tendencia en la que Guatemala y El Salvador presentan simultáneamente mayor carga de cuidados y menor acceso a internet \(51\.45% y 47\.62%, respectivamente\) dentro de la muestra analizada, lo que sugiere posibles contextos de mayor vulnerabilidad relativa\.

## 2\.3\. Vulnerabilidad estructural

Mide la combinación de carga de cuidados y brecha digital, como un indicador agregado del nivel de vulnerabilidad estructural\. Este es el índice combinado del proyecto, que integra ambas variables en un puntaje de riesgo para identificar los contextos más críticos por país\.

Cálculo: \(carga de cuidados \+ brecha digital\) / 2
Es decir, se calcula el valor medio entre ambos problemas\. Los valores más altos indican mayor vulnerabilidad estructural, al combinar mayor carga de cuidados y menor acceso digital\.

```
df_tiempo_internet["vulnerabilidad"] = (
    df_tiempo_internet["carga_cuidados"] + df_tiempo_internet["brecha_digital"]
) / 2
```

```
# Revisamos las primeras filas
df_tiempo_internet[["pais", "carga_cuidados", "brecha_digital", "vulnerabilidad"]].head(10)
```

```
# Ranking de vulnerabilidad estructural por país
df_tiempo_internet[["pais", "vulnerabilidad"]].sort_values(
    by="vulnerabilidad",
    ascending=False
)
```

Conclusión preliminar: Los resultados de este índice combinado refuerzan, a nivel general, las conclusiones preliminares obtenidas en los análisis previos, evidenciando patrones consistentes entre carga de cuidados y brecha digital\. En conjunto, este indicador permite sintetizar en una sola medida dos dimensiones estructurales clave del problema, facilitando la comparación entre países y la identificación de tendencias generales dentro de la muestra\. Asimismo, resulta clave para identificar contextos prioritarios de intervención, ya que permite reconocer los países con situaciones más críticas en términos de vulnerabilidad estructural\.

## 2\.4\. Peso poblacional \(edad activa\)

Calcula la proporción de mujeres en edad activa en cada país, como aproximación al volumen de población potencialmente expuesta a las condiciones de trabajo y brecha digital\. Este indicador se incorpora para dimensionar la magnitud real del problema en cada país, ya que no solo importa la intensidad de la vulnerabilidad estructural, sino también qué porcentaje de mujeres en la edad mencionada se ven afectadas\. De este modo, permite contextualizar el impacto potencial de las condiciones analizadas y aportar una mirada de escala poblacional al análisis\.

La elección del rango etario 15–64 no implica que el resto de los grupos no sea relevante, sino que responde al recorte analítico del proyecto, el cual se enmarca dentro de la categoría “trabajo remunerado”\. En este sentido, se prioriza la denominada “población en edad de trabajar”, definida por la OECD \(Organización para la Cooperación y el Desarrollo Económico\) como la proporción de personas entre 15 y 64 años respecto del total de la población\.

Cálculo: \(suma de miles de personas en los grupos de edad 15–64\) / total de mujeres del país

Los valores más altos indican una mayor proporción de mujeres en edad activa dentro del país\.

```
# Filtramos el dataset y sólo dejamos las filas correspondientes al grupo etario de 15-64 años
df_activa = df_poblacion[df_poblacion["grupos_edad"].isin([
    "15-19","20-24","25-29","30-34","35-39",
    "40-44","45-49","50-54","55-59","60-64"
])]

# Cálculo de población total por país (todas las edades)
pob_total = df_poblacion.groupby("pais")["miles_personas_a_mitad_anio"].sum()

# Cálculo de población en edad activa por país
pob_activa = df_activa.groupby("pais")["miles_personas_a_mitad_anio"].sum()

# Cálculo del peso poblacional de la edad activa
peso_poblacional_edad_activa = (pob_activa / pob_total) * 100
```

```
# Ranking de países de mayor a menor proporción de mujeres en edad activa
peso_poblacional_edad_activa.sort_values(ascending=False)
```

Conclusión preliminar: Se observa una distribución relativamente homogénea de la población femenina en edad activa entre los países analizados, con variaciones acotadas entre el 63% y el 68%, lo que sugiere que las diferencias en el impacto del índice estarán principalmente determinadas por el grado de vulnerabilidad estructural y no por la composición demográfica\.
No obstante, se trata en todos los casos de más de la mitad de la población femenina, lo que refuerza la relevancia de considerar este aspecto en el análisis de la problemática\.

## 2\.5\. Riesgo poblacional

El riesgo poblacional permite identificar qué países concentran mayor nivel de criticidad dentro del análisis, combinando la vulnerabilidad estructural con el peso de la población en edad activa\. De esta manera, no solo se considera la intensidad de las condiciones de desigualdad en términos de carga de cuidados y brecha digital, sino también cuántas mujeres están potencialmente expuestas a estas condiciones en cada país\.

Cálculo: vulnerabilidad estructural \* proporción de población en edad activa

Esto permite obtener una medida integrada del impacto potencial del problema en cada contexto nacional\. En este sentido, valores más altos indican países donde la combinación entre condiciones estructurales adversas y mayor exposición poblacional en edad activa genera escenarios más críticos, lo que resulta clave para la identificación de prioridades de intervención\.

```
# ---------------------------------------------------
# DATASET DE RIESGO POBLACIONAL (NIVEL PAÍS)
# ---------------------------------------------------

# Partimos del dataset de indicadores a nivel país
df_riesgo = df_tiempo_internet[["pais", "vulnerabilidad"]].copy()

# Agregamos el peso poblacional de la edad activa
# (proporción de mujeres en edad 15-64 dentro del total)
df_riesgo = df_riesgo.merge(
    peso_poblacional_edad_activa.rename("peso_poblacional_edad_activa"),
    on="pais",
    how="left"
)

# Cálculo del riesgo poblacional:
# combinación de vulnerabilidad estructural y peso poblacional
df_riesgo["riesgo_poblacional"] = (
    df_riesgo["vulnerabilidad"] * df_riesgo["peso_poblacional_edad_activa"]
) / 100
```

```
# Ranking de riesgo poblacional

df_riesgo = df_riesgo.sort_values(
    by="riesgo_poblacional",
    ascending=False
)

df_riesgo
```

Conclusión preliminar: Se observa que Guatemala lidera el ranking de riesgo poblacional, seguido por El Salvador\. Esto indica que, dentro de la muestra analizada, estos países no solo presentan mayores niveles de vulnerabilidad estructural, sino que además dicha situación se ve amplificada por el peso relativo de su población en edad activa\. En conjunto, estos resultados refuerzan la identificación de estos países como los contextos más críticos dentro del análisis, al concentrar simultáneamente mayor intensidad de desigualdades y mayor exposición poblacional\.
